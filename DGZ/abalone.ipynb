```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVR

param_distributions = {
    'C': [0.1, 1, 10, 100],
    'epsilon': [0.01, 0.1, 0.2, 0.5, 1],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'gamma': ['scale', 'auto']
}

svr_model = SVR()

random_search = RandomizedSearchCV(
    svr_model,
    param_distributions=param_distributions,
    n_iter=20,
    cv=5,
    scoring='neg_mean_squared_error',
    random_state=0,
    n_jobs=-1
)

random_search.fit(X_train, y_train)

best_params = random_search.best_params_
best_score = -random_search.best_score_

print("Best parameters found: ", best_params)
print("Best cross-validation accuracy: ", best_score)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

svr_model = SVR(kernel='rbf', gamma='scale', C=100, epsilon=0.5)
svr_model.fit(X_train, y_train)

y_pred = svr_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MSE: {mse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"RÂ²: {r2:.2f}")
```

```python
from sklearn.model_selection import cross_val_score

# K-fold CV (K=5)
mse_scores = -cross_val_score(svr_model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
r2_scores = cross_val_score(svr_model, X_train, y_train, cv=5, scoring='r2')

print(f"Mean MSE: {mse_scores.mean():.2f}")
print(f"MSE Standard Deviation: {mse_scores.std():.2f}")
print("\n")
print(f"Mean RÂ² Score: {r2_scores.mean():.2f}")
print(f"RÂ² Score Standard Deviation: {r2_scores.std():.2f}")
```

```python
plt.figure(figsize=(14, 6))

# MSE
plt.subplot(1, 2, 1)
plt.plot(range(1, 6), mse_scores, marker='o', linestyle='-', color='b', label='MSE per Fold')
plt.axhline(y=mse_scores.mean(), color='r', linestyle='--', label=f'Mean MSE: {mse_scores.mean():.2f}')
plt.title('MSE Scores Across 10 Folds')
plt.xlabel('Fold')
plt.ylabel('MSE')
plt.ylim([0, 10])
plt.legend()
plt.grid(True)

# R2
plt.subplot(1, 2, 2)
plt.plot(range(1, 6), r2_scores, marker='o', linestyle='-', color='g', label='RÂ² per Fold')
plt.axhline(y=r2_scores.mean(), color='r', linestyle='--', label=f'Mean RÂ²: {r2_scores.mean():.2f}')
plt.title('RÂ² Scores Across 10 Folds')
plt.xlabel('Fold')
plt.ylabel('RÂ²')
plt.ylim([0, 1])
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
